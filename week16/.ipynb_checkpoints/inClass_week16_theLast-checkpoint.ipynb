```python
filename = '/Users/jillnaiman/Downloads/othello.txt'
```

```python
# read it in!
with open(filename) as f:
    text = f.read()
```

```python
#print(text)
```

Split text into words:

```python
myWords = text.split()
```

```python
myWords
```

```python
# use a dictionary object to count words
wordsDict = {}
for word in myWords:
    if word in wordsDict: # if already in the dictionary
        wordsDict[word] += 1
    else:
        wordsDict[word] = 1
```

```python
wordsDict
```

First before making a quick histogram, let's sort in order of decreasing occurance:

```python
import operator
```

```python
wordsDict = sorted(wordsDict.items(), key=operator.itemgetter(1))
# this is basically sorting our dictionary object by the value
```

```python
wordsDict[::-1]
```

```python
# put sorted back into dictionary
wordsDict = dict(wordsDict[::-1])
```

```python
wordsDict
```

Start plotting already!

```python
n = 50 # plot only the 50 most popular words

wordsPlot = {k: wordsDict[k] for k in list(wordsDict)[:n]}
```

```python
wordsPlot
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

# let's add the actual words to our x-axis labels
ax.set_xticks(range(len(wordsPlot)))
ax.set_xticklabels(list(wordsPlot.keys()))

plt.xticks(rotation=90)

plt.show()
```

```python
len(wordsDict) # how many unique words?
```

We have a lot of words!  But some are "stop words" or frequent but not (usually) useful words AND we have "repeats" of similar words ("the" vs "The").

```python
# us nltk package to grab the stopwords
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

Make a new dictionary from our text without these stopwords:

```python
less_words = {}
for w, num in wordsDict.items():
    #print(w, num) # pairs of words, count of occurances
    if w.lower() not in stop_words:
        less_words[w] = num
```

```python
less_words
```

```python
len(less_words)
```

Remake our histogram plot using the dictionary that has been cleaned of stop words:

```python
n = 50 # 50 most popular

wordsPlot = {k: less_words[k] for k in list(less_words)[:n]}

fig, ax = plt.subplots(figsize=(10,3))

ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

# adding in labels for our individual words
ax.set_xticks(range(len(wordsPlot)))
ax.set_xticklabels(list(wordsPlot.keys()))

plt.xticks(rotation=90)

plt.show()
```

Next steps:
1. take into account the fact that there can be multiple versions of the same word (pant, pants)
1. take out stopwords
1. puncation


```python
word_tokens = nltk.word_tokenize(" ".join(myWords))
```

```python
word_tokens
```

```python
len(word_tokens) # tokenized corpus (full corpus)
```

```python
# take out stop words
less_words = [w for w in word_tokens if not w.lower() in stop_words]
```

```python
less_words
```

```python
len(less_words)
```

Take out trailing punctuation/individual punctuation marks:

```python
string.punctuation
```

```python
no_punc_words = []
for word in less_words:
    #print(word)
    if word[-1] in string.punctuation: # check for trailing punctuation
        word = word[:-1] # take off any trailing punctation
    if len(word) > 0: # I have something left, i.e. my original word was not JUST punctuation
        no_punc_words.append(word)
```

```python
no_punc_words
```

Note: could have done take out puncutation first, and then do take out stopwords

```python
wordsDict = {}
for word in no_punc_words:
    if word in wordsDict:
        wordsDict[word] += 1
    else:
        wordsDict[word] = 1
```

```python
wordsDict
```

Re-make our histogram:

```python
wordsDict = sorted(wordsDict.items(), key=operator.itemgetter(1))
```

```python
wordsDict = dict(wordsDict[::-1])
```

```python
wordsDict # stand to take out a few more stop words!
```

```python
n = 50 # 50 most popular

wordsPlot = {k: wordsDict[k] for k in list(wordsDict)[:n]}

fig, ax = plt.subplots(figsize=(10,3))

ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

# adding in labels for our individual words
ax.set_xticks(range(len(wordsPlot)))
ax.set_xticklabels(list(wordsPlot.keys()))

plt.xticks(rotation=90)

plt.show()
```

Before continuing to clean our data, let's make a wordcloud:

```python
wordcloud = WordCloud(colormap='viridis').generate_from_frequencies(wordsDict)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.imshow(wordcloud)

plt.show()
```

One quick adding of a word to our stop list (i.e. including more things to take out):

```python
stop_words
```

```python
stop_words2 = stop_words.copy()
```

```python
if 'OTHELLO' not in stop_words2:
    stop_words2.add('OTHELLO')
```

Redo the counting of words:

```python
no_oth_words = []
for word in no_punc_words:
    if word not in stop_words2:
        no_oth_words.append(word)
```

```python
wordsDict = {}
for word in no_oth_words:
    if word in wordsDict:
        wordsDict[word] += 1
    else:
        wordsDict[word] = 1
```

```python
wordcloud = WordCloud(colormap='viridis').generate_from_frequencies(wordsDict)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.imshow(wordcloud)

plt.show()
```
