```python
filename = '/Users/jnaiman/Downloads/facebook_combined_sm000090_000010.txt' 
#filename = '/Users/jnaiman/Downloads/facebook_combined_sm000030_000000.txt' 
```

```python
network = pd.read_csv(filename, sep=' ', names=['ind1', 'ind2'])
network = network.drop_duplicates()
```

```python
network
```

```python
import numpy as np
```

```python
node_data = []
link_data = []
color_data = []

u_nodes = np.unique(np.append(network['ind1'],network['ind2'])) # unique nodes in the network (unique people)

for un in u_nodes:
    node_data.append({'label':str(un), "shape":"circle"})
    color_data.append('blue')
    
for iu,un in enumerate(u_nodes): # loop over unique nodes
    target_ids = network.loc[network['ind1']==un]
    #print(target_ids)
    #print(' ')
    tnodes = np.unique(target_ids['ind2'].values)
    #print(un,tnodes)
    #print(' ')
    for t in tnodes: # link all associate target nodes for this un, unique node
        tind = np.where(t == u_nodes)[0][0] # what is the order of this target in u_nodes
        link_data.append({'source':iu, 'target':tind})
```

```python
graph = bqplot.Graph(node_data = node_data, link_data = link_data, colors=color_data)

graph.charge = -50 # default is -1000, -100 means things are closer together
graph.directed = False
graph.link_type = 'line'
graph.link_distance = 50

fig = bqplot.Figure(marks=[graph])
fig.layout.min_width = '1000px'
fig.layout.min_height = '900px'
fig
```
