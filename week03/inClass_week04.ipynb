```python
# step 2: put into function
def make_plot(style):
    #with plt.style.context("ggplot"):
    with plt.style.context(style):
        fig, ax = plt.subplots(figsize=(8,3))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year')
        ax.set_ylabel("Square Footage")
        ax.set_yscale('log')
        ax.legend()
    plt.show()
```

```python
# let's make a widget for looking at colormaps!
# step 1: hard-code

colormap_name = 'jet' # not a great colormap, btw!

fig, ax = plt.subplots(figsize=(8,0.8))
col_map = plt.get_cmap(colormap_name)
#col_map
# default for drawing a colorbar (not worry about if you don't want to!)
matplotlib.colorbar.ColorbarBase(ax,cmap=col_map,orientation='horizontal')
ax.xaxis.set_visible(False) # turn off x-axis
plt.show()
```

```python
# step 2: function it!
def make_colorbar(colormap_name):
    fig, ax = plt.subplots(figsize=(8,0.8))
    col_map = plt.get_cmap(colormap_name)
    # default for drawing a colorbar (not worry about if you don't want to!)
    matplotlib.colorbar.ColorbarBase(ax,cmap=col_map,orientation='horizontal')
    ax.xaxis.set_visible(False) # turn off x-axis
    plt.show()
```

```python
make_colorbar("jet")
```

```python
make_colorbar("Blues")
```

```python
# step 3: decorate our function
@ipywidgets.interact(colormap_name=plt.colormaps())
def make_colorbar(colormap_name):
    fig, ax = plt.subplots(figsize=(8,0.8))
    col_map = plt.get_cmap(colormap_name)
    # default for drawing a colorbar (not worry about if you don't want to!)
    matplotlib.colorbar.ColorbarBase(ax,cmap=col_map,orientation='horizontal')
    ax.xaxis.set_visible(False) # turn off x-axis
    plt.show()
```

Note: we didn't use the "with" statement here -- not changing matplotlib defaults

```python
# step 2, take 2
def michigan_plot(colormap_name, take_log=True):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10)
    else:
        norm = matplotlib.colors.Normalize()
    plt.imshow(michigan, cmap=colormap_name, norm = norm)
    plt.colorbar()
    plt.clim(-352,352)
    plt.show()
```
