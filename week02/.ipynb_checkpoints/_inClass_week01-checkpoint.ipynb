```python
# GDP (gross domestic product over the time period in question)
gdp = [14233.2, 14422.3, 14569.7, 14685.3, 14668.4, 14813.0, 14843.0, 14549.9, 14383.9]
```

```python
gdp
```

```python
# let's plot out the GDP as a function of time
plt.plot(time, gdp)
plt.xlabel('Time in format YearMonthDay')
plt.ylabel('GDP in Billions')
```

```python
plt.plot(time, gdp, 'o-')
plt.xlabel('Time in format YearMonthDay')
plt.ylabel('GDP in Billions')
```

We are going to use the datetime package to help us format our x-axis in a "familiar" way -- as dates instead of large numbers.

```python
time
```

```python
from datetime import datetime
```

```python
date = [] #empty list to hold our dates

for t in time: # loop though each object in list
    date.append( datetime.strptime(str(t), '%Y%m%d'))
```

```python
date
```

```python
plt.plot(date, gdp, 'o-')
plt.xlabel('Date')
plt.ylabel('GDP in Billions')
```

## Reading in data the hardway

```python
def converter(v):
    return datetime.strptime(v.decode("ascii"), '%Y-%m-%d')
```

```python
# where I've downloaded data
myFredFile = '/Users/jillnaiman/Downloads/GDP.csv'
```

I'm going to use numpy to do some data conversions:

```python
import numpy as np
```

```python
myFredType = np.dtype([("date", datetime), ("gdp", np.float64)])
```

```python
with open(myFredFile, "r") as f:
    data = np.loadtxt(f, skiprows=1, delimiter=",", 
                      converters={0: converter}, 
                      dtype=myFredType)
```

```python
data
```

```python
plt.plot(data['date'], data['gdp'], 'o-')
plt.xlabel('Date')
plt.ylabel('GDP in Billions')
```
