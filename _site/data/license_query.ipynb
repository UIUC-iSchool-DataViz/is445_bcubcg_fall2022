Getting data from: https://data.illinois.gov/dataset/professional-licensing/resource/beef79ab-b679-4b0a-960a-05c218a619ba

```python
save_file = '/Users/jnaiman/Downloads/tmp/license_records_1000000.csv'
```

```python
import urllib
import json
import pandas as pd
```

```python
#url = 'https://data.illinois.gov/api/3/action/datastore_search?resource_id=beef79ab-b679-4b0a-960a-05c218a619ba&limit=5&q=title:jones'  
#url = 'https://data.illinois.gov/api/3/action/datastore_search?resource_id=beef79ab-b679-4b0a-960a-05c218a619ba&limit=5'  
#url = 'https://data.illinois.gov/api/3/action/datastore_search?resource_id=beef79ab-b679-4b0a-960a-05c218a619ba&limit=100000'  
url = 'https://data.illinois.gov/api/3/action/datastore_search?resource_id=beef79ab-b679-4b0a-960a-05c218a619ba&limit=1000000'  
fileobj = urllib.request.urlopen(url)
data = fileobj.read().decode('utf-8')
```

```python
j = json.loads(data)
```

```python
df = pd.DataFrame(j['result']['records'])
```
